```r
library('magrittr')
library('tibble')
library('dplyr')
library('tidyr')
library('ggplot2')
library('RColorBrewer')
library('readr')
```

```r
conf_flow <- 
    read_tsv(
        'data.tsv',
        col_types=cols(
            `Sample:`=col_character(),
            `Debead`=col_factor(levels=c('Day 1', 'Day 2', 'Day 3')),
            `Sample`=col_character(),
            `Replicate`=col_factor(levels=c('R1', 'R2', 'R3', 'R4')),
            `Seed`=col_factor(levels=c('2000000', '1000000', '500000', '250000', '125000')),
            .default=col_double()
        )
    )

conf_flow %>% head
```

```r
seed_labeler <-
    c(
        `2000000`="Seed: 2,000,000 cells",
        `1000000`="Seed: 1,000,000 cells",
        `500000`="Seed: 500,000 cells",
        `250000`="Seed: 250,000 cells"
    )

conf_flow %>%
    mutate(
        `CM count`=
            `Count` *
            4 * # 4 because used 1/4 of the sample for this run
            `Dilution` *  # which was further diluted to 1/2
            (`Live | Freq. of Parent`/100.0) *
            (`Live/CCR7+ CD45RO+ | Freq. of Parent`/100.0)
    ) %>%
    filter(`Seed` != '125000') %>%
    ggplot(aes(x=`Debead`, y=`CM count`, color=`Debead`)) +
        geom_boxplot(alpha=0.1) +
        geom_point(pos=position_dodge(width=.7)) +
        facet_wrap(~`Seed`, labeller=as_labeller(seed_labeler)) +
        theme(axis.text.x = element_text(angle=45, hjust=1)) +
        ylab('Central Memory Count') +
        xlab('Debeading Day')
```
