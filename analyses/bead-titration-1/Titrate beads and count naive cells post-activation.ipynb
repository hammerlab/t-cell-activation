```r
library('magrittr')
library('tibble')
library('dplyr')
library('tidyr')
library('ggplot2')
library('RColorBrewer')
library('readr')
```

```r
d35_flow <- 
    read_tsv(
        'donor35-titration.tsv',
        col_types=cols(
            `Sample:`=col_character(),
            `Bead/Cell Ratio`=col_factor(levels=c('0', '0.03', '0.1', '0.3', '1', '3')),
            `Replicate`=col_factor(levels=c('R1', 'R2', 'R3', 'R4')),
            `Day`=col_factor(levels=c('Day 3', 'Day 5')),
            `Count`=col_number(),
            .default=col_double()
        )
    )

d35_flow %>% head
```

```r
d35_flow %>%
    select(
        `Sample:`:`Live | Freq. of Parent`,
        `Naive: CCR7+ CD45RO-`=`Live/CCR7+ CD45RO- | Freq. of Parent`,
        `CM: CCR7+ CD45RO+`=`Live/CCR7+ CD45RO+ | Freq. of Parent`,
        `EM: CCR7- CD45RO+`=`Live/CCR7- CD45RO+ | Freq. of Parent`,
        `EMRA: CCR7- CD45RO-`=`Live/CCR7- CD45RO- | Freq. of Parent`
    ) %>%
    gather(
        key='Population',
        value='Population Frequency',
        factor_key=TRUE,
        `Naive: CCR7+ CD45RO-`:`EMRA: CCR7- CD45RO-`
    ) %>%
    group_by(`Bead/Cell Ratio`, `Day`, `Population`) %>%
    summarize(
        `Avg_frequency`=mean(`Population Frequency`),
        `Min_frequency`=`Avg_frequency` - sd(`Population Frequency`),
        `Max_frequency`=`Avg_frequency` + sd(`Population Frequency`)
    ) %>%
    ggplot(aes(x=`Population`, y=`Avg_frequency`, fill=`Bead/Cell Ratio`)) +
        geom_col(position="dodge") +
        geom_errorbar(
            aes(
                ymin=`Min_frequency`,
                ymax=`Max_frequency`
            ),
            position=position_dodge(width=.9),
            alpha=0.5
        ) +
        ylim(0, 100) +
        ylab('Percent population (%)') +
        theme(axis.text.x=element_text(angle=75, hjust=1)) +
        ggtitle('Frequencies of T cell subtypes across activation conditions') +
        facet_wrap(~`Day`, ncol=2)
```

```r
norm_by_0 <- function(x) {
    return(x / x[1])
}

d35_flow %>%
    select(
        `Sample:`:`Live | Freq. of Parent`,
        `Naive: CCR7+ CD45RO-`=`Live/CCR7+ CD45RO- | Freq. of Parent`
    ) %>%
    mutate(`Raw count`=`Count` * `Live | Freq. of Parent` * `Naive: CCR7+ CD45RO-`) %>%
    group_by(`Replicate`, `Day`) %>%
    mutate(`Absolute count`=1 - norm_by_0(`Raw count`)) %>%
    group_by(`Bead/Cell Ratio`, `Day`) %>%
    summarize(
        `Avg_count`=mean(`Absolute count`),
        `Min_count`=`Avg_count` - (sd(`Absolute count`)/3),
        `Max_count`=`Avg_count` + (sd(`Absolute count`)/3)
    ) %>%
    ggplot(aes(x=`Bead/Cell Ratio`, y=`Avg_count`, fill=`Day`)) +
        geom_col(position="dodge") +
        geom_errorbar(
            aes(
                ymin=`Min_count`,
                ymax=`Max_count`
            ),
            position=position_dodge(width=.9),
            alpha=.5,
            width=.5
        ) +
        ylab("Fraction activated") +
        ggtitle("Fraction of naive cells activated")
```
