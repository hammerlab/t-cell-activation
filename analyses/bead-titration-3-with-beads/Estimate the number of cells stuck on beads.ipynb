```r
library('magrittr')
library('tibble')
library('dplyr')
library('tidyr')
library('ggplot2')
library('RColorBrewer')
library('readr')
```

```r
bead_flow <- 
    read_tsv(
        'data.tsv',
         col_types=cols(
             `Sample:`=col_character(),
             `Day`=col_factor(levels=c('Day 1', 'Day 2', 'Day 3', 'Day 4')),
             `Bead_to_cell`=col_factor(levels=c('3', '1', '0.3', '0.1', '0.03', '0')),
             `Donor`=col_factor(levels=c('Donor 52', 'Donor 53', 'Donor 54', 'Donor 55')),
             .default=col_double()
         )
    )

bead_flow %>% head
```

```r
bead_flow %>%
    mutate(`Cells stuck on beads`=`Count` * `Bead+Cells | Freq. of Parent`) %>%
    ggplot(aes(x=`Bead_to_cell`, y=`Cells stuck on beads`, color=`Bead_to_cell`)) +
        geom_boxplot(alpha=.1) +
        geom_jitter() +
        ylab('Number of cells stuck on beads') +
        facet_wrap(~`Day`, ncol=2) +
        xlab('Bead to cell ratio')
```
